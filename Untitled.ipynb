```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cap = cv2.VideoCapture(1)
ret, frame = cap.read()
cap.release()
```

```python
plt.figure(figsize=(18, 16))

#gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='Greys_r')
plt.show()
```

```python
csobelx=cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=3)
csobely=cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=3)
plt.figure()
plt.imshow(csobelx,cmap='Greys_r')
plt.show()
plt.figure()
plt.imshow(csobely,cmap='Greys_r')
plt.show()

csobel=cv2.add(cv2.abs(csobelx)+cv2.abs(csobely))
```
